```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = x^2
def f(x):
    return x**2

# Define the derivative function f'(x) = 2x
def df(x):
    return 2*x
```

```python
# Generate x values from -5 to 5
x = np.linspace(-5, 5, 100)
y = f(x)
y_derivative = df(x)
```

```python
# Plot the function f(x)
plt.figure(figsize = (10, 5))
plt.plot(x, y, label = r'$f(x) = x^2$', color = 'blue')

# Plot the derivative f'(x)
plt.plot(x, y_derivative, label = r"$df'(x) = 2x$", color = 'red', linestyle = 'dashed')

# Highlight key points
plt.scatter([0], [f(0)], color = 'green', marker = 'o', s = 100, label = "Minimum (0,0)")
plt.scatter([3, -3], [f(3), f(-3)], color = 'purple', marker = 'o', s= 100, label = "Sample Points")

# Add labels
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.title("Function and its Derivative")

# Show the plot
plt.show()
```

```python
# Implementing basic gradient descent
def gradient_descent(start_x, learning_rate, iterations):
    x = start_x   # starting from the initial point
    history = []  # storing values for visualization

    for _ in range(iterations):
        gradient = df(x)  # Computing derivative
        x = x - learning_rate * gradient  # updating x
        history.append(x)   # storing for plotting
    
    return x, history  # Final value of x and trajectory

# Running the gradient descent
final_x, trajectory = gradient_descent(start_x = 5, learning_rate = 0.1, iterations = 10)

print("Final x after Gradient Descent:", final_x)
print("Trajectory:", trajectory)
```

```python
# Plotting the trajectory for the above gradient descent

# Function and its derivative
def f(x):
    return x**2

# Generating x values for the plot
x_values = np.linspace(-6, 6, 400)
y_values = f(x_values)

# Running the gradient descent
final_x, trajectory = gradient_descent(start_x = 5, learning_rate = 0.1, iterations = 10)

# Computing the function values at trajectory points
trajectory_y = [f(x) for x in trajectory]

# Plotting the function f(x) = x^2
plt.figure(figsize = (8, 5))
plt.plot(x_values, y_values, label = r'$f(x) = x^2$', color = 'blue')

# Plotting the trajectory points
plt.scatter(trajectory, trajectory_y, color = 'red', marker = 'o', label = "Gradient Descent Steps")

# Highlighting starting and final points
plt.scatter([5], [f(5)], color = 'green', s = 100, label = "Start Point (x = 5)")
plt.scatter([final_x], [f(final_x)], color = 'purple', s = 100, label = "FInal Point")

# Labels and titles
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent on f(x) = x^2")
plt.legend()
plt.grid()
plt.show()
```

Based on all the observations done with different iterations, instead of running 100 iterations for different LR, adding a new stopping condition.

```python
# New stopping condition: Stop if the change in 'x' is less than a small threshold  = 10^-6

def gradient_descent(start_x, learning_rate, max_iterations=100, epsilon=1e-6):
    x = start_x   # Start at initial point
    history = []  # Store trajectory for visualization

    for _ in range(max_iterations):
        gradient = df(x)  # Compute derivative
        new_x = x - learning_rate * gradient  # Update x

        # Store for plotting
        history.append(new_x)

        # Check stopping condition
        if abs(new_x - x) < epsilon:  
            break  # Stop if change is very small
        
        x = new_x  # Update x for next iteration

    return x, history  # Final value of x and trajectory
```

```python
# Running the gradient descent with different learning rates
final_x, trajectory = gradient_descent(start_x=-3, learning_rate=0.01)
final_x1, trajectory_1 = gradient_descent(start_x=-3, learning_rate=0.05)
final_x2, trajectory_2 = gradient_descent(start_x=-3, learning_rate=0.2)

print("Final x for LR 0.01:", final_x, "Iterations:", len(trajectory))
print("Final x for LR 0.05:", final_x1, "Iterations:", len(trajectory_1))
print("Final x for LR 0.2:", final_x2, "Iterations:", len(trajectory_2))
```
